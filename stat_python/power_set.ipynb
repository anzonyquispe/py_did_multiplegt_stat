```python
from itertools import combinations


def power_set(items):
    """
    Return all non-empty subsets of `items`.

    Parameters
    ----------
    items : list
        A list of items.

    Returns
    -------
    out : list[list]
        List of subsets (each subset is a list), excluding the empty set.
        Ordered by subset size (1..k), and within each size, lexicographic by input order.
    """
    items = list(items)
    out = []
    for r in range(1, len(items) + 1):
        for combo in combinations(items, r):
            out.append(list(combo))
    return out
```
